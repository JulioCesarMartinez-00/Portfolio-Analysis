# Introduction to Data Visualization with Matplotlib 

```python
# Import the matplotlib.pyplot submodule and name it plt
import pandas as pd
import matplotlib.pyplot as plt
```

```python
seattle_weather = pd.read_csv('./Datasets/seattle_weather.csv')
austin_weather = pd.read_csv('./Datasets/austin_weather.csv')
```

```python
# groupby per month
seattle_weather = seattle_weather.groupby(['DATE']).mean('MLY-PRCP-NORMAL').reset_index()
austin_weather = austin_weather.groupby(['DATE']).mean('MLY-PRCP-NORMAL').reset_index()
```

```python
# Create a Figure and an Axes with plt.subplots
fig, ax = plt.subplots()

# Plot MLY-PRCP-NORMAL from seattle_weather against the MONTH
ax.plot(seattle_weather["DATE"], seattle_weather["MLY-PRCP-NORMAL"])

# Plot MLY-PRCP-NORMAL from austin_weather against MONTH
ax.plot(austin_weather['DATE'], austin_weather['MLY-PRCP-NORMAL'])

# Call the show function
plt.show()
```

```python
# customaize your plot
fig, ax = plt.subplots()

# Plot MLY-PRCP-NORMAL from seattle_weather against the MONTH
ax.plot(seattle_weather["DATE"], seattle_weather["MLY-PRCP-NORMAL"],marker="o",linestyle="--",color="b")

# Plot MLY-PRCP-NORMAL from austin_weather against MONTH
ax.plot(austin_weather['DATE'], austin_weather['MLY-PRCP-NORMAL'],marker="v",linestyle="--",color="r")

# Call the show function
plt.show()
```

# Labels and title

```python
# customaize your plot
fig, ax = plt.subplots()

# Plot MLY-PRCP-NORMAL from seattle_weather against the MONTH
ax.plot(seattle_weather["DATE"], seattle_weather["MLY-PRCP-NORMAL"],marker="o",linestyle="--",color="b")

# Plot MLY-PRCP-NORMAL from austin_weather against MONTH
ax.plot(austin_weather['DATE'], austin_weather['MLY-PRCP-NORMAL'],marker="v",linestyle="--",color="r")

# Customize the x-axis label
ax.set_xlabel("Time (months)")

# Customize the y-axis label
ax.set_ylabel("Precipitation (inches)")

# Add the title
ax.set_title("Weather patterns in Austin and Seattle")
# Call the show function

plt.show()
```

# Multipe subplots

```python
# customaize your plot
fig, ax = plt.subplots(2,1)

ax[0].plot(seattle_weather["DATE"], seattle_weather["MLY-PRCP-NORMAL"],marker="o",linestyle="--",color="b")
ax[1].plot(austin_weather['DATE'], austin_weather['MLY-PRCP-NORMAL'],marker="v",linestyle="--",color="r")

ax[1].set_xlabel("Time (months)")
ax[0].set_ylabel("Precipitation (inches)")
ax[1].set_ylabel("Precipitation (inches)")
ax[0].set_title("Weather patterns in Austin and Seattle")

plt.show()
```
